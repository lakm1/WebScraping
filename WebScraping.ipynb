Now doing the same steps but for all the products by collecting the all the pages links + product links
then collecting the data from each with the same steps done before

````python
# Collect all page links
all_pages_links = []
current_link = link

while True:
    all_pages_links.append(current_link)

    request2 = requests.get(current_link)
    soup2 = BeautifulSoup(request2.text, "html.parser")

    next_btn = soup2.select_one("a.next.page-numbers") # the next button
    if next_btn and next_btn.has_attr("href"):
        current_link = next_btn["href"]
    else:
        break
````

````python
print("Number of pages found >>", len(all_pages_links))
print(all_pages_links)
````

````python
product_links = []

for url in all_pages_links:
  # Request & Soup each page a
    request_page = requests.get(url,verify=False)
    soup_page = BeautifulSoup(request_page.text, "html.parser")


    products_list = soup_page.find("ul", class_="products")
    products = products_list.find_all("li", class_="product")

    for product in products:
        a = product.find("a", class_="woocommerce-LoopProduct-link")
        if a and a.has_attr("href"):
          link = a["href"]
          product_links.append(link)
````

````python
print("Total product links >>", len(product_links))
print(product_links)
````

````python
# Loop over product links and extract info

rows = []

for link in product_links:
    request_p = requests.get(link,verify=False)
    soup_p = BeautifulSoup(request_p.text, "html.parser")

    # title
    t = soup_p.find("h1", class_="product_title")
    title = t.get_text(strip=True) if t else None

    # image
    img = soup_p.find("img", class_="wp-post-image")
    image = img["src"] if img and img.has_attr("src") else None

    # price
    p = soup_p.find("p", class_= "price")
    price = p.get_text(" ").split()[1] if p else None # took the price without £


    # availability
    a = soup_p.find("p", class_= "stock in-stock")
    available = a.get_text(" ").split()[0] if a else None

    # categories
    p = soup_p.find("span", class_= "posted_in")
    cat_links = p.find_all("a") if p else []
    cats = [c.get_text(strip=True) for c in cat_links] if cat_links else [] # in list, cause there is multiple


    # tags
    tags = soup_p.find("span", class_= "tagged_as")
    tag_links = tags.find_all("a")
    tags = [t.get_text(strip=True) for t in tag_links] if tag_links else []

    # description
    d = soup_p.find("div", class_= "woocommerce-product-details__short-description")
    desc = d.get_text(" ", strip=True) if d else None

    rows.append({ "title": title, "image": image,
                 "price": price, "availability": available,
                  "categories": ", ".join(cats),
                  "tags": ", ".join(tags),
                  "description": desc })



````

````python
rows
````

````python
# adding the rows in Pandas DataFrame
df = pd.DataFrame(rows)
df
````

````python
# Saving the data in csv file
df.to_csv("items_data.csv", index=False)
````
